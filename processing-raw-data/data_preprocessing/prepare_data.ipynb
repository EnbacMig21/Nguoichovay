Đây là bảng thông tin các khoản vay tại ngân hàng.

```python
default_df = pd.read_csv('Tình trạng vỡ nợ của khoản vay.csv')
```

```python
default_df
```

Đây là tình trạng vỡ nợ của các khoản vay đó.

Chúng ta có 3 bảng từ đó phải gộp thành 1 bảng để dễ dàng cho việc phân tích.

```python
temp_df = pd.merge(loan_df, borrower_df, on='Customer_ID', how='inner')
final_df = pd.merge(temp_df, default_df, on='Loan_ID', how='inner')
final_df
```

```python
final_df.info()
```

Có thể thấy dữ liệu này không có khuyết thiếu dữ liệu (do tất cả đều là non-null)

## Xóa "Loan_ID" & "Customer_ID"

Ta thấy hai cột "Loan_ID" và Customer_ID" dùng để liên kết các bảng với nhau nên sau khi gộp lại thì không còn cần dùng đến nữa có thể loại bỏ.

```python
final_df = final_df.drop(['Loan_ID','Customer_ID'], axis = 1)
final_df
```

#### Cột "Loan_Term" đang có kiểu dữ liệu str nên cần chuyển qua dạng int64 để tính toán.

```python
final_df['Loan_Term'] = final_df['Loan_Term'].str.replace(' months', '').astype('int64')
final_df
```

```python
final_df['Gender'] = final_df['Gender'].replace({'Male': 1, 'Female': 0}).astype('int64')
final_df
```
